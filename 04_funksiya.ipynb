# 1) Chiziqli va kvadrat funksiyalar garafiklari

$ y = 3x-5 $ funksiya grafigini chizing.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(-5,5)

y = 3*x - 5

plt.plot(x,y)
plt.show()
```

$ y = 3x - 2 $ va $ y = x^2 - 3x + 2 $ funksiyalar grafigini bitta chizmada chizing.

```python
x = np.linspace(-5,7)

y1 = 3*x - 2
y2 = x**2 - 3*x + 2

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

# 2) Ko'rsatkichli funksiyaning grafigi

```python
x = np.linspace(-3,4)

y3 = 2**x - 4

plt.plot(x,y3)
plt.show()
```

# 3) Logarifmik funksiyaning grafigi

$ y = log_2 (x+4) $ funksiya grafigini chizing.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,100)

y4 = np.log2(x+4)

plt.plot(x,y4)

plt.show()
```

# 4) Trigonometrik va teskari trigonometrik funksiyalarning grafiklari

$ y=sin2x $ va $ y=2cos \frac{x}{2} $ garafiklarini bitta chizmada tasvirlang.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi,2*np.pi)

y5 = np.sin(2*x)
y6 = np.cos(x/2)

plt.plot(x,y5)
plt.plot(x,y6)

plt.show()
```

$ y = 2arctgx - arcctgx $ funksiya grafigini chizing.

```python
x = np.linspace(-5,5)

y7 = 2*np.arctan(x) + np.arctan(1/x) 

plt.plot(x,y6)
plt.plot(x,y7)
plt.show()
```

# 6) Uch o'lchamli grafiklar chizish

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import*
```

```python
f = lambda x,y: np.sin(x) + np.cos(y)

fig=plt.figure(figsize=(10,10), dpi=100)

x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)

x,y = np.meshgrid(x,y)
g = f(x,y)

ax=fig.gca(projection='3d')
ax.plot_surface(x,y,g, cmap=cm.coolwarm)

plt.show()
```

# 7) Funksiya grafigiga ishlov berish  

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib.inline()
```

```python
x = np.linspace(-3,3)

y9 = 2*np.sin(3*x)
y99 = np.cos(x)

plt.figure(figsize=(7,4), dpi=100)

plt.title("Bu sarlavha")
plt.xlabel("x sonlar")
plt.ylabel("$ y = 2sin3x $")

plt.xlim(-3,3)
plt.ylim(-3,3)

plt.plot(x,y9,'o-', color='red', lw=3, ms=10, label="y=2sin3x" )
plt.plot(x,y99,'--', color='blue', lw=2, ms=5, label="y=cosx" )

plt.legend(loc="upper right")

plt.grid()
plt.show()
```

# 8)  Subplot, axes plot

```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib.pyplot import*
```

```python
x1 = np.linspace(-2*np.pi,2*np.pi)
x2 = x3 = np.linspace(-5,5)

y1 = np.sin(x1)
y2 = np.tan(x2)
y3 = x3**2/2
y4 = np.cos(x1/2)
y5 = x2**2 - 3*x2
y6 = np.arctan(x1)


plt.figure(1)
plt.figure(figsize=(15,5), dpi=100)

plt.subplots_adjust(wspace=.4, hspace=.4)

subplot(2,3,1)
plt.plot(x1,y1,'--', color="red", label="y=sinx")
plt.legend()

subplot(2,3,2)
plt.plot(x2,y2, '--', color="green", label="y=tg(x/2)")
plt.legend(loc='upper right')


subplot(2,3,3)
plt.plot(x2,y3, '--', color="blue", label='$y=x^2 /2$')
plt.legend()

subplot(2,3,4)
plt.plot(x1,y4,'--', color="red", label="$y=cos \frac{x}{2}$")
plt.legend()

subplot(2,3,5)
plt.plot(x2,y5, '--', color="green", label="$y=x^2-3x$")
plt.legend(loc='upper right')


subplot(2,3,6)
plt.plot(x1,y6, '--', color="blue", label='$y=arctgx$')
plt.legend()
```

# 9) Parametrik plot va contour plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x,y: np.sin(x) - np.cos(y)
    
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

x,y = np.meshgrid(x,y)
g = f(x,y)

# plt.contour(x,y,g)    # faqat chiziqlarni o'zini ko'rsatadi
plt.contourf(x,y,g)     # chiziqlar ichkasirisini turli rang bilan to'ldiradi
plt.show()
```

$ z = q^{4} + 3 q^{2} r^{2} + r^{4} - 2 r^{2} \sqrt{q^{2} + r^{2}} \quad $ ni contourplot da chizing.

```python
import matplotlib.pyplot as plt
import numpy as np 
import sympy as sp
```

```python
q,r = sp.symbols('q,r')
```

```python
z = r**4 + 3*q**2*r**2 + q**4 - 2*r**2*sp.sqrt(r**2+q**2)
z
```

```python
r = np.linspace(0,2)
q = np.linspace(0,0.7)
r,q = np.meshgrid(r,q)

z = r**4 + 3*q**2*r**2 + q**4 - 2*r**2*(r**2+q**2)**0.5

plt.grid(True)
plt.xlabel("Q/M")
plt.ylabel("r/M")

plt.title("Q/M vs r/M")

plt.contourf(q, r, z,5)

plt.colorbar();
plt.show()
```

Agar $\theta \in [-2\pi ; 2 \pi] \quad va \quad z \in [-3;3] \quad $ bo'lsa $ r = z^2 + 1 \quad x = r \cdot sin \theta \quad y = r \cdot cos \theta \quad $ z $\quad$ ni parametrik plot yordamida chizing. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ax = plt.figure(figsize=(5,5), dpi=100).add_subplot(projection='3d')

θ = np.linspace(-2 * np.pi, 2 * np.pi, 500)
z = np.linspace(-3, 3, 500)

r = z**2 + 1

x = r * np.sin(2*θ)
y = r * np.cos(θ)

ax.plot(x, y, z, label='parametrik plot')
ax.legend()

plt.show()
```
